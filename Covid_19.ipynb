```python
days=df[df['Current Status']=='Hospitalized'].groupby(['Month','Day'])['Num Cases'].sum()
days
```

```python
len(days)
```

```python
## generate an array of 365 days
a=np.arange(len(days))
a
```

```python
days.values
```

```python
b=days.values
plt.scatter(a,b)
plt.show()
```

```python
## a is 1d array for ML algo it should be in 2d array
```

```python
a=a.reshape(-1,1)
```

```python
a
```

```python
from sklearn.linear_model import LinearRegression

r=LinearRegression()
r.fit(a,b)
```

```python
b[364]
```

```python
r.predict([[365]])
```

```python
b_predict = r.predict(a)
```

```python
## predicted by machine
b_predict
```

```python
## our actual value
b
```

```python
plt.scatter(a,b)
plt.plot(a,b_predict)
plt.show()
```

```python
r.score(a,b)*100
```
